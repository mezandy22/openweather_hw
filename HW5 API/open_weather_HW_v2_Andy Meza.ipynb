```python
# Dependencies
import requests
import json
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from citipy import citipy


# Weather
api_key = "f2ad267e6b392fa25066799d9964e648"

#beginning of url
url = "http://api.openweathermap.org/data/2.5/weather?"

#units of measurement
units = "imperial"

# Build partial query URL. Cities will go on the end during the request
query_url = url + "appid=" + api_key + "&units=" + units + "&q="



```

```python
#get random numbers generated for longitude and altitude
#latitude ranges form -90 to +90
#longitude ranges from -180 to +180
import random

test_latitude = random.uniform(-90, 90)
test_latitude = float("{0:.2f}".format(test_latitude))

test_longitude =  random.uniform(-180, 180)
test_longitude = float("{0:.2f}".format(test_longitude))

print(test_latitude)
print(test_longitude)
```

```python
##place random generated lat and long into the citipy code

test_rand_city = citipy.nearest_city(test_latitude, test_longitude)
test_rand_city

test_rand_city_name = test_rand_city.city_name
test_rand_city_name
```

```python
##get x amount of random cities, using citipy, by changing the range in the loop cities 

city_list = []

for i in range(1, 800):
    latitude = random.uniform(-90, 90)
    latitude = float("{0:.2f}".format(latitude))

    longitude =  random.uniform(-180, 180)
    longitude = float("{0:.2f}".format(longitude))
    
    #using citipy with lat and long used above
    rand_city = citipy.nearest_city(latitude, longitude)
    
    #getting actual name from citipy 
    rand_city_name = rand_city.city_name
    rand_city_name = rand_city_name.replace(" ", "%20") ### people suggested I import urllib for a dynamic url. What do they mean?
    city_list.append(rand_city_name)
    print("________")
    city_weather = query_url + rand_city_name
    print("Beginning Data Retrieval")
    print("__________________________")
    print("processing record {} | {}".format(i, rand_city_name))
    print(city_weather)
    i += 1

# print(city_list)
    
```

```python
# Request Data using the city list we made using citipy
city_weather = [requests.get(query_url + rand_city_name).json() for rand_city_name in city_list]
# print(json.dumps(city_weather, indent=2))
```

```python
import pandas as pd
# Extract interesting data from responses
lat_data = [data.get("coord").get("lat") for data in city_weather]
lng_data = [data.get("coord").get("lon") for data in city_weather]
temp_data = [data.get("main").get("temp_max") for data in city_weather]
humidity_data = [data.get("main").get("humidity") for data in city_weather]
cloudiness_data = [data.get("clouds").get("all") for data in city_weather]
wind_speed_data = [data.get("wind").get("speed") for data in city_weather]
city_data = [data.get("name") for data in city_weather]
country_data = [data.get("sys").get("country") for data in city_weather]
date_data = [data.get("dt") for data in city_weather] 


weather_data = {"temp_max": temp_data, "lat": lat_data, "lng": lng_data,"humidity":humidity_data, "date":date_data,
                    "cloudiness":cloudiness_data, "wind speed": wind_speed_data, "city": city_data, "country": country_data}

weather_data = pd.DataFrame(weather_data)

```
